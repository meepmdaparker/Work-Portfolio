```python
def reconfiguration_engine(df1:pd.DataFrame, df2:pd.DataFrame):
    reconfiguration_list = df1[['vehicle_id', 'config_string']]
    reconfiguration_list['best_match_score'] = ''

    #when using a for loop on a pandas dataframe -> for i in df1 -> this iterates over column names, not the rows, which is why we need to use iterrows
    for index, non_compliant_vehicle in df1.iterrows():
        # Filter df2 based on matching 'model_powertrain', 'model_year', and 'country'
        filtered_df2 = df2.loc[(df2['model_powertrain'] == non_compliant_vehicle['rep_vehicle_model_powertrain']) &
                            (df2['model_year'] == non_compliant_vehicle['model_year']) &
                            (df2['country'] == non_compliant_vehicle['country']) &
                            (df2['paint_id'] == non_compliant_vehicle['build_items_exp_exterior_id'])
                            ]

        # Extract the config strings from the filtered df2
        compliant_choices = filtered_df2['config_string'].tolist()

        # Use fuzzywuzzy to find the best match
        best_match = process.extractOne(non_compliant_vehicle['config_string'], compliant_choices)

        # Assign the best match and score to the current row in reconfiguration_list
        if best_match is not None:
            reconfiguration_list.loc[index, 'best_match_config_string'] = best_match[0]
            reconfiguration_list.loc[index, 'best_match_score'] = best_match[1]
        else:
            reconfiguration_list.loc[index, 'best_match_config_string'] = 'Best Matching config_string could not be found; Does compliant config_string with exact model_powertrain, model_year, or country exist'
            reconfiguration_list.loc[index, 'best_match_score'] = 0
    return reconfiguration_list




new_output1 = reconfiguration_engine(rep_vehicles_df_final_filtered.copy(), inventory_compliant_recommendation_population_df.copy())




#is there a faster way to do this by applying a hashmap?
#for example a dictionary of key (non-compliant vehicleID) and value (non-compliant vehicle_id's config_string best_match score and best_match config_string)



```

```python
from fuzzywuzzy import process
import pandas as pd

def faster_reconfiguration_engine(df1: pd.DataFrame, df2: pd.DataFrame):
    reconfiguration_list = df1[['vehicle_id', 'config_string']].copy()
    reconfiguration_list['best_match_config_string'] = None
    reconfiguration_list['best_match_score'] = None

    # Create a dictionary to store compliant choices in df2, keyed by (model_powertrain, model_year, country)
    compliant_choices_dict = df2.groupby(['model_powertrain', 'model_year', 'country'])['config_string'].apply(list).to_dict()

    for index, non_compliant_vehicle in df1.iterrows():
        key = (non_compliant_vehicle['rep_vehicle_model_powertrain'],
               non_compliant_vehicle['model_year'],
               non_compliant_vehicle['country'])

        compliant_choices = compliant_choices_dict.get(key, [])

        best_match = process.extractOne(non_compliant_vehicle['config_string'], compliant_choices)

        if best_match is not None:
            reconfiguration_list.loc[index, 'best_match_config_string'] = best_match[0]
            reconfiguration_list.loc[index, 'best_match_score'] = best_match[1]
        else:
            reconfiguration_list.loc[index, 'best_match_config_string'] = 'Best Matching config_string could not be found; Does compliant config_string with exact model_powertrain, model_year, or country exist'
            reconfiguration_list.loc[index, 'best_match_score'] = 0

    return reconfiguration_list

new_output_hashmap = faster_reconfiguration_engine(rep_vehicles_df_final_filtered.copy(), inventory_compliant_recommendation_population_df.copy())
```

```python
import pandas as pd
import re

def splitting_config_string_attributes_engine_enhanced(df: pd.DataFrame):
    prefix_to_column = {
        'EXP': 'original_exterior_id',
        'INT': 'original_interior_id',
        'WHL': 'original_wheel_id',
        'TON': 'original_tonneau_id',
        'ACTBDG': 'original_badging_id',
        'ROOF': 'original_roof_id',
        'AUD': 'original_audio_id',
        'UTL': 'original_utility_id',
    }

    df['original_model_year'] = None
    df['original_country_id'] = None
    df['original_model_powertrain'] = None
    for prefix in prefix_to_column.values():  # Create the new marketing attribute columns
        df[prefix] = None

    for index, row in df.iterrows():
        config_string = row['config_string']
        if isinstance(config_string, str):
            parts = config_string.split('_')
            if len(parts) > 0:
                df.loc[index, 'original_model_year'] = parts[0]
            if len(parts) > 1:
                country_code = parts[1]
                if country_code == 'US':
                    df.loc[index, 'original_country_id'] = 'RGN-USA'
                elif country_code == 'CA':
                    df.loc[index, 'original_country_id'] = 'RGN-CAN'
                else:
                    df.loc[index, 'original_country_id'] = f'RGN-{country_code}' # Handle other countries if needed
            if len(parts) > 2:
                df.loc[index, 'original_model_powertrain'] = '_'.join(parts[2].split(' ')[0:3]) # Take the first 2 or 3 words

            for part in parts:
                for prefix, column_name in prefix_to_column.items():
                    if part.startswith(prefix + '-'):
                        df.loc[index, column_name] = part
                        break

    df['original_atp_id'] = pd.NA
    df.loc[df['original_wheel_id'].isin(['WHL-0AD', 'WHL-0A1', 'WHL-0DN', 'WHL-0DD']), 'original_atp_id'] = 'ATP-001'
    df.loc[~df['original_wheel_id'].isin(['WHL-0AD', 'WHL-0A1', 'WHL-0DN', 'WHL-0DD']), 'original_atp_id'] = 'ATP-NULL'

    return df

# Example usage (assuming 'new_output1' is your DataFrame)
new_output_with_split_data_enhanced = splitting_config_string_attributes_engine_enhanced(new_output_with_split_data_enhanced.copy())
```

```python
import pandas as pd

def compare_and_clear_marketing_attributes(df: pd.DataFrame):
    attribute_pairs = {
        'marketing_exterior': 'original_exterior_id',
        'marketing_interior': 'original_interior_id',
        'marketing_wheel': 'original_wheel_id',
        'marketing_tonneau': 'original_tonneau_id',
        'marketing_badging': 'original_badging_id',
        'marketing_roof': 'original_roof_id',
        'marketing_audio': 'original_audio_id',
        'marketing_utility': 'original_utility_id',
    }

    for marketing_col, original_col in attribute_pairs.items():
        if marketing_col in df.columns and original_col in df.columns:
            df.loc[df[marketing_col] == df[original_col], marketing_col] = pd.NA

    return df

# Example usage (assuming 'df_with_marketing' is your DataFrame)
# Make sure your DataFrame has the 'marketing_*' and 'original_*_id' columns
# df_with_marketing = ...

processed_df = compare_and_clear_marketing_attributes(new_output_with_split_data_enhanced.copy())
```
