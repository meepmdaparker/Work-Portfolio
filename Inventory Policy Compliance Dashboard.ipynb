````python
import typing
import hextoolkit
import pandas

def sql_cell(
    source: str,
    data_connection_name: typing.Optional[str] = None,
    cast_decimals: bool = True,
) -> "pandas.DataFrame":
    """
    Args:
        source (str): SQL string to run against pandas DataFrame or Data Connection.
            When run against a pandas DataFrame, the `from_item` in the `FROM` clause
            is the name of the Python variable that is assigned to the pandas DataFrame

            ```
            df1 = pandas.DataFrame({...})
            sql_cell("select * from 'df1'")
            ```

        data_connection_name (Optional[str], optional): Data Connection to run `source` against.
            If not provided, assumes SQL `source` is to be run on DataFrame. Defaults to None.

        cast_decimals (bool, optional): Whether or not to cast Decimal types to floats. This
            parameter is exposed to the Hex UI as `Infer numeric datatypes`.
            Defaults to True.

    Returns:
        pandas.DataFrame: The result of running `source` as if it was in a SQL Cell
    """
    if data_connection_name:
        hex_data_connection = hextoolkit.get_data_connection(data_connection_name)
        return hex_data_connection.query(source, cast_decimals=cast_decimals)
    return hextoolkit.query_dataframes(source, cast_decimals=cast_decimals)


dataframe_3 = vehicle_kpi_df
````
